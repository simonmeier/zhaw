```python
import pandas as pd
```

```python
df = pd.read_csv('investments_VC.csv',encoding = 'unicode_escape')
df.head()
```

```python
cats = df['category_list']
```

```python
cats.dropna(inplace=True)
```

```python
#Welche Kategorien gibt es
cats_set = {''}
for str in cats:
    try:
     cats_indiv = str.split('|')
    except:
       print(str)
    for cat in cats_indiv:
       cats_set.add(cat)
cats_set
```
